```python

import numpy as np 
import pandas as pd

```

```python
df=pd.read_csv("train_transaction.csv")
test=pd.read_csv("test_transaction.csv")
df2=pd.read_csv("train_identity.csv")
test2=pd.read_csv("test_identity.csv")
```

```python
df.head()
```

```python
df3=pd.merge(df,df2,on="TransactionID",how="left")
df3.head()
```

```python
test3=pd.merge(test,test2,on="TransactionID",how="left")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
listt=[np.int32,np.int64,np.float32,np.float64,np.float16,np.int16]
for feat in df3.columns:
    #if type(df[feat][0])==np.order
    le=LabelEncoder()
    
    if not type(df3[feat][0]) in listt:
         original=df3[feat]
         mask=df3[feat].isnull()
         #df3[feat]=df3[feat].fillna(type(df3[feat][0])(-1000))
         
         df3[feat]=df3[feat].astype(str)
         df3[feat] = le.fit_transform(df3[feat])
         df3[feat][mask]=np.nan
         
        
        
   
```

```python
listt=[np.int32,np.int64,np.float32,np.float64,np.float16,np.int16]
for feat in test3.columns:
    #if type(df[feat][0])==np.order
    le=LabelEncoder()
    
    if not type(test3[feat][0]) in listt:
         original=test3[feat]
         mask=test3[feat].isnull()
         #df3[feat]=df3[feat].fillna(type(df3[feat][0])(-1000))
         
         test3[feat]=test3[feat].astype(str)
         test3[feat] = le.fit_transform(test3[feat])
         test3[feat][mask]=np.nan
         
        
   
```

```python

target=df3.isFraud
X=df3.drop("isFraud",axis=1)
#X=torch.from_numpy(X)
#target=torch.from_numpy()
#to_var(X)
#to_var(target)
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,target,test_size=.1)
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(train_x, label=train_y)
d_val  =lgb.Dataset(test_x,label=test_y)
params = {}
params['learning_rate'] = 0.005
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 200
params['min_data'] = 100
params["metric"]='auc'
params['max_depth'] = -1
params['n_estimators'] = 2000
clf = lgb.train(params, d_train, valid_sets=[d_train,d_val] ,verbose_eval=50,early_stopping_rounds=300)
```

```python

y_pred=clf.predict(test3)
```

```python
sub = pd.read_csv('sample_submission.csv', index_col = 'TransactionID')
sub['isFraud']=y_pred
sub.to_csv('submission.csv')
sub.head()
```
